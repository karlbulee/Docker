# Windows Terminal Anaconda 추가

1. Windows Terminal 실행(MS 스토어)
2. 설정 클릭
3. 좌측 하단의 Json 파일 열기 클릭
4. list 부분 찾기
5. 다은 코드 추가  
{
          "guid": "{7번 코드}",
          "hidden": false,
          "name": "Anaconda",
          "commandline": "9번 코드"
        }
6. 새로운 PowerShell 에서 [guid]::NewGuid() 코드 실행
7. 결과 {7번 코드}에 넣기 ex) "guild": "{dcf45ee6-6ad8-4728-8368-4372ad4aaa56}"
8. Anaconda Prompt 속성 열어서 경로 복사 
9. commandline에 붙여넣기 ex) "commandline": "\"%windir%\\System32\\cmd.exe \"/K\" C:\\ProgramData\\Anaconda3\\Scripts\\activate.bat C:\\ProgramData\\Anaconda3"
10. Windows Terminal 재실행
