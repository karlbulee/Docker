# Windows terminal에 Git Bash 추가하기

1. Windows Terminal 실행
2. 설정 클릭
3. 프로필에서 새 프로필 추가 > 새 빈 프로필 클릭
4. 
이름에 Git Bash, 
명령줄에 GIt Bash 경로(이 때 새로운 창으로 열리지 않도록 bin\bash.exe로 수정), 
아이콘에 Git Bash 아이콘 경로 입력 후 저장
5. 확인
