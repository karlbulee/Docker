# Dockerfile 작성하기

- Dockerfile을 작성하는 일은 그렇게 어려운 일이 아니다
- mlflow git에서 사용한 Dockerfile 내용은 다음과 같다
- 쉘스크립트로 만들거나 VS Code를 이용하면 훨씬 간단하다
- VS Code를 사용할 때는 docker extension을 추가해야 한다

```python
# Dockerfile 위치 경로
C:\Users\canmanmo\Desktop\mlflow_custom\Docker
    
    
# Dockerfile 내용
FROM condaforge/miniforge3

RUN pip install mlflow>=1.0 \
    && pip install azure-storage-blob==12.3.0 \
    && pip install numpy==1.21.2 \
    && pip install scipy \
    && pip install pandas==1.3.3 \
    && pip install scikit-learn==0.24.2 \
    && pip install cloudpickle
```

## 성공 결과

# Docker Image 생성

- Dockerfile은 이미지를 작성할 수 있는 설정 파일임
- Image 이름은 MLproject file과 매칭되어야 한다(선 image 후 MLproject file)

```python
# "-t <생성할 이미지명>:<태그명> <Dockerfile 위치>"
docker build -t mlflow-docker-example -f Dockerfile .
```

## 성공 결과

# Docker Container 생성

```python
docker run mlflow-docker-example
```
