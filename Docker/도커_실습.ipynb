# 컨테이너 사용 방법

- 컨테이너를 다루는 모든 명령은 docker 명령어로 시작

```python
docker ~ 
```

**커맨드**
  - ocker 명령어 뒤에 오는 '무엇을' '어떻게'에 해당하는 부분
  - 상위 커맨드 : 무엇을. 대상의 종류가 들어가고 딱 12종류뿐 
  - 하위 커맨드 : 어떻게
  - 대상 : 컨테이너명 또는 이미지명 등 구체적인 이름 
  - docker 명령 뒤로 '무엇을', '어떻게', '대상' 순으로 지정
    - docker 커맨드 대상
  - 역사적 이유로 container를 붙이지 않아도 실행 가능한 명령어 있음 
    - docker run

```python
# container : 상위 커맨드
# run : 하위 커맨드
# penguin : 대상 
docker container run penguin 

docker image pull penguin 

docker container start penguin 
```

**옵션과 인자**
  - 옵션 : 커맨드에 세세한 설정 지정
  - 인자 : 대상에 전달할 값 지정
  - docker 상위 커맨드 하위 커맨드 옵션 대상 인자
  - 자주 쓰이는 옵션이나 인자 한정돼 있으므로 기억 
  - '-' 또는 '--'으로 시작(둘 중에 무엇 사용 기준 없음)

```python
docker container run -d penguin --mode=1
```

**정리**
docker 커맨드 (옵션) 대상 (인자)
