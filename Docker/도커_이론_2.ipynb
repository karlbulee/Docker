# 이미지와 컨테이너

- 컨테이너로도 이미지 만들 수 있음
- 이미 만든 컨테이너에 손을 대서 컨테이너를 개조함 
- 개조된 컨테이너로부터 이미지 만들고 나면 새로 만든 이미지 사용해 개조된 컨테이너 여러 개 만들기 가능
- 컨테이너로부터 만든 이미지 이용하면 다른 환경에서 원래의 컨테이너 복원 가능
- 도커 원칙 : '한 컨테이너에 한 프로그램'

## 도커 허브와 이미지

- 초보자가 컨테이너의 이미지 밑바닥부터 만들어 낼 일 거의 없음
- 도커 허브 : 공식적으로 운영되는 도커 레지스트리(도커 이미지 배포 서비스)
- 공개된 컨테이너 이미지가 모여 있는 곳
- https://hub.docker.com/
- 운영체제(비슷한 것)만 들어 있는 이미지부터, 여러 가지 소프트웨어가 함께 포함된 것까지 다양한 이미지 제공
- 공식 이미지(도커 직접 배포 or 소프트웨어 개발 및 관리 기업이나 조직에서 제공)
- 커스텀 이미지 : 최소한의 요소가 담긴 이미지에 필요한 소프트웨어 추가 설치

# 도커 컨테이너 생애주기

- 컨테이너는 쓰고 버리는 일회용품 같은 것 
- 컨테이너 하나를 업데이트하면서 계속 사용하기보다는 업데이트된 소프트웨어가 들어있는 새로운 컨테이너를 사용하는 것이 좋음
- 오래된 컨테이너를 버리고 새로운 이미지로부터 새로운 컨테이너 만들어 갈아타기
- 컨테이너 생애주기 : 컨테이너 만들기 => 실행하기 => 종료하기 => 폐기 
- 컨테이너 폐기하면 안에서 편집했던 파일은 사라지므로 보통 도커가 설치된 물리적 서버(호스트)의 디스크를 마운트해 여기에 데이터를 저장함

# 도커 장단점

*환경 격리*
- 독립된 환경 : 여러 개의 컨테이너 띄우기. 똑같은 애플리케이션 여러 개 띄우기
- 이미지 생성 : 교체 및 업데이트 쉬움
- 커널 없음 : 가볍고 배포판 원하는 것 사용 가능

1. 장점
- 한 대의 물리 서버에 여러 대의 서버 띄울 수 있음 
- 서버 관리 용이
- 서버 고수가 아니어도 다루기 쉬움

2. 단점
- 리눅스용 소프트웨어밖에 지원하지 않음
- 호스트 서버에 문제 생기면 모든 컨테이너에 영향 미침

## 도커의 용도

- 팀원 모두에게 동일한 개발환경 제공하기(=동일한 환경 여러 개 만들기)
- 새로운 버전을 개발환경에서 테스트한 후 운영환경에 적용할 때 컨테이너 활용
- 동일한 서버가 여러 대 필요한 경우 

# 도커 사용

*도커 환경 구축*
- 리눅스 컴퓨터에서 도커 사용
- 가성 머신이나 렌탈 환경에 도커 설치하고 윈도우나 macOS 통해 사용
- 윈도우용/macOS용 도커 사용

## 윈도우용/macOS용 도커 사용

- 윈도우와 macOS에서는 리눅스 운영체제가 들어있는 패키지(도커 데스크톱) 사용
- 사용자가 가상화 환경이나 리눅스 운영체제 신경 쓰지 않아도 됨
- 도커가 전용으로 사용하는 숨겨진 운영체제기 때문에 운영체제 두 개여도 괜찮음
- 윈도우에서는 Hyper-V(윈도우용 가상환경)가 활성 상태여야 함

## WSL2

- WSL2(Windows Subsystem for Linux 2) : 윈도우에서 리눅스 소프트웨어 실행하기 위한 기능
- 도커 제작사에서 만든 리눅스 운영체제를 사용하는 버전과 마이크로소프트에섬 ㅏㄴ든 WSL2를 사용하는 버전 있음

## 도커 실행 조건

- 64비트 운영체제에서만 동작
- 윈도우 설정 > 시스템 > 정보
