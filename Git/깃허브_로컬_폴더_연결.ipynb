# 깃허브 로컬 폴더 연결하기

- 깃 -> 로컬
1. Git Bash에서 로컬 폴더 위치로 이동 ex) cd Desktop/ML
2. git remote add origin https://github.com/karlbulee/ML.git
3. git pull origin main

- 로컬 -> 깃
1. git add --all
2. git commit -m "커밋 로그"
3. git push origin main
