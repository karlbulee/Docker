# 대용량 파일 에러 대처법

- github은 100M 이상의 파일을 올리려고 하면 오류가 발생

# 해결방법

1. https://rtyley.github.io/bfg-repo-cleaner/ 파일 다운로드
2. 해당 파일 repo 옮기기
3. java -jar bfg-1.14.0.jar --strip-blobs-bigger-than 100M 명령어 실행
