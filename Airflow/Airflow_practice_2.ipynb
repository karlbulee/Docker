# Airflow

## Airflow 용 가상환경

```python
conda create -n airflow python=3.7
conda activate airflow
```

## Airflow 설치 코드

```python
pip install apache-airflow
airflow db init
airflow users create --username admin --password admin --firstname Anonymous --lastname Admin --role Admin -e admin@example.org
mkdir ~/airflow/dags/
cp download_rocket_launches.py ~/airflow/dags/
airflow webserver
airflow scheduler
```

### Error
- [ERROR] Can't connect to ('0.0.0.0', 8080)
